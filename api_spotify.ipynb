# **3. Obtenemos info de las 10 canciones top de los 10 artistas top**

```python
# Ids de artistas globales (top 10)
artist_ids = ["06HL4z0CvFAxyc27GXpf02","4q3ewBCX7sLwd24euuV69X","1Xyo4u8uXC1ZmMpatF05PJ","3TVXtAsR1Inumwj472S9r4","12GqGscKJx3aE4t07u7eVZ",
              "2LRoIwlKmHjgvigdNGBHNo","0Y5tJX1MQlPlqiwlOH1tJY","7tYKF4w9nC0nq9CsPZTHyP","790FomKkXshlbRYZFtlgla","00FQb4jTyendYWaN8pK0wa"]
```

```python
# Lista de DataFrames
df_list = []

# Generar un DataFrame para cada artista
for artist_id in artist_ids:
    # Por defecto arroja las 10 mejores canciones
    top_tracks = sp.artist_top_tracks(artist_id)
    top_tracks_df = pd.DataFrame(top_tracks)
    top_tracks_df["artist_id"] = artist_id
    df_list.append(top_tracks_df)

# Concatenar todos los DataFrames de la lista verticalmente
artists_top_tracks_df = pd.concat(df_list, ignore_index=True)
```

```python
artists_top_tracks_df.head(1)
```

```python
# Explotamos la columna "tracks"
artists_top_tracks_df_n = pd.json_normalize(artists_top_tracks_df["tracks"])
artists_top_tracks_df_n.head(1)
```

```python
# Fusionar el nuevo dataframe explotado y la columna "artist_id"
artists_top_tracks_df=pd.merge(artists_top_tracks_df["artist_id"],artists_top_tracks_df_n,left_index=True,right_index=True)
artists_top_tracks_df.head(3)
```

```python
# Crear un nuevo DataFrame con los valores necesarios
artists_top_tracks_df = pd.DataFrame({
    "artist_id": artists_top_tracks_df["artist_id"],
    "duration_ms": artists_top_tracks_df["duration_ms"],
    "track_name": artists_top_tracks_df["name"],
    "popularity": artists_top_tracks_df["popularity"],
    "preview_url": artists_top_tracks_df["preview_url"],
    "album.images": artists_top_tracks_df["album.images"].apply(lambda x: x[0]["url"]),
    "album_name": artists_top_tracks_df["album.name"],
    "release_date": artists_top_tracks_df["album.release_date"]
})

# Imprimir el nuevo DataFrame
artists_top_tracks_df
```

```python
# Escribimos nuestro dataframe
artists_top_tracks_df.to_excel("artist_top_tracks_df.xlsx", index=False)
```
